```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
from sklearn.metrics import log_loss
```

```python
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('../data/porto-seguro-safe-driver-prediction/train.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
X = data.drop(columns=['id','target'])
y = data['target']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.5)
X_train,X_train_lr,y_train,y_train_lr = train_test_split(X_train,y_train,test_size=0.5)
```

```python
n_estimator = 10
```

```python
# GBDT
grd = GradientBoostingClassifier(n_estimators=n_estimator)
grd.fit(X_train, y_train)
# OneHot
grd_enc = OneHotEncoder(categories='auto')

temp = grd.apply(X_train)
np.set_printoptions(threshold=np.inf)
grd_enc.fit(grd.apply(X_train)[:, :, 0])
#print(grd_enc.get_feature_names()) # 
# OneHotLR
grd_lm = LogisticRegression(solver='lbfgs', max_iter=1000)
grd_lm.fit(grd_enc.transform(grd.apply(X_train_lr)[:, :, 0]), y_train_lr)
# LR
y_pred_grd_lm = grd_lm.predict_proba(grd_enc.transform(grd.apply(X_test)[:, :, 0]))[:, 1]
fpr_grd_lm, tpr_grd_lm, _ = roc_curve(y_test, y_pred_grd_lm)

```

```python
log_loss(y_test,y_pred_grd_lm)
```

```python
p = sum(y_test)/len(y_test)
p
```

```python
ne = log_loss(y_test,y_pred_grd_lm)/(-1*(p*np.log(p)+(1-p)*np.log(1-p)))
print('ne =',ne)
```

```python
test = pd.read_csv('../data/porto-seguro-safe-driver-prediction/test.csv')
```

```python
pred = grd_lm.predict_proba(grd_enc.transform(grd.apply(test.drop(columns=['id']))[:, :, 0]))
```
