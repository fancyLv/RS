```python
import numpy as np
```

```python
a = np.array([[0,   0, 0, 1/3, 0,   0],
              [1/4, 0, 0, 0,   1/2, 0],
              [0,   1, 0, 1/3, 1/2, 0],
              [1/4, 0, 0, 0,   0,   1],
              [1/4, 0, 1, 1/3, 0,   0],
              [1/4, 0, 0, 0,   0,   0]])
```

```python
b = np.array([1/6,1/6,1/6,1/6,1/6,1/6])
w = b
```

```python
# 简化模型
def work(a,w):
    for i in range(100):
        w = np.dot(a,w)
        print(w)
    return w
```

```python
# 随机模型
def random_work(a,w,n):
    d = 0.85
    for i in range(100):
        w = (1-d)/n + d * np.dot(a,w)
        print(w)
    return w
```

```python
work(a,w)
```

```python
random_work(a,w,6)
```
